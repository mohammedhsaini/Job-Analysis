```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import WebDriverException
from urllib.parse import urlparse
from bs4 import BeautifulSoup as bs
from itertools import zip_longest
import csv
```

```python
job_titles =[]
companies = []
locations = []
job_types =[]
time_posted_list = []
links = []
urls = []
skills = []
experiences= []
education = []
salaries =[]
pages_url = []
requierments = []
job_categories= []
sizes = []
sectors =[]
```

```python
for i in range(0, 13):
    page_num = str(i)
    URL = f'https://wuzzuf.net/search/jobs/?a=hpb&q=data%20analyst&start={i}'
    pages_url.append(URL)
```

```python
pages_url
```

```python
URL1 = "https://wuzzuf.net"
for URL in pages_url:
    try:
        driver = webdriver.Chrome('C:/Users/lenovo/Downloads/chromedriver')
        driver.get(URL)
        html = driver.page_source
        soup = bs(html, 'html.parser')
        containers = soup.find_all("div", {"class": "css-1gatmva e1v1l3u10"})
        for container in containers:
            link = container.find("a",{"class": "css-o171kl"})['href']
            links.append(link)
            jtitle = container.find("h2",{'class': 'css-m604qf'}).text.strip()
            job_titles.append(jtitle)
            company = container.find("a",{'class': 'css-17s97q8'}).text.strip()
            companies.append(company)
            location = container.find("span",{'class': 'css-5wys0k'}).text.strip()
            locations.append(location)
            job_type = container.find("span",{'class': 'css-1ve4b75 eoyjyou0'}).text.strip()
            job_types.append(job_type)
    except WebDriverException as e:
        print(f"Selenium Error: {e}")
    finally:
        driver.quit()
```

```python
for link in links:
    link = link.replace("data analyst", "dataanalyst")
    url = URL1+link
    urls.append(url)
```

```python

for url in urls:
    try:
        cdriver = webdriver.Chrome('C:/Users/lenovo/Downloads/chromedriver')
        cdriver.get(url)
        html1 = cdriver.page_source
        s = bs(html1, 'html.parser')

        sector_element = s.find("span", {"class": "css-xilyze"})
        if sector_element:
            sector = sector_element.text.strip()
            if sector:
                sectors.append(sector)
            else:
                print('Sector found but has no text')
        else:
            print('Sector not found')

        size_element = s.find("span", {"class": "css-nhiaul"})
        if size_element:
            size = size_element.text.strip()
            if size:
                sizes.append(size)
            else:
                print('Size found but has no text')
        else:
            print('Size not found')

        container1 = s.find_all("section", {"class": "css-3kx5e2"})
        for con1 in container1:
            reqs = con1.find_all("span", {"class": "css-4xky9y"})
            requirements_concatenated = ', '.join(req.text.strip() for req in reqs)
            if requirements_concatenated:
                requierments.append(requirements_concatenated)

        container2 = s.find_all("div", {"class": "css-s2o0yh"})
        for con2 in container2:
            elements = con2.find_all("span", {"class": "css-158icaa"})
            skills_concatenated = ', '.join(el.text.strip() for el in elements)
            if skills_concatenated:
                skills.append(skills_concatenated)

    except WebDriverException as e:
        print(f"Selenium Error: {e}")
    finally:
        cdriver.quit()

```

```python
file_list = [job_titles,locations,requierments,skills,companies,sizes,sectors,urls]
exported = zip_longest(*file_list)
```

```python
with open(r"C:/Users/lenovo/python learning/indeed Job data.csv", "w",newline = '',encoding = "utf-8") as my_file:
    wr = csv.writer(my_file,delimiter=',', lineterminator='\n')
    wr.writerow(["Job Title","Location","Requierments","Skills","Company Name","Company Size","Sector","URL"])
    wr.writerows(exported)
```
